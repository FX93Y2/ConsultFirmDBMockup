```python
import pandas as pd
file_path = '/Users/FangYan/Desktop/Team4-Mock ConsultingRDB/dataset/Consultant_TitleHistory_Denorm.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
consultant_table = data[['ConsultantID', 'FirstName', 'LastName', 'Email']].drop_duplicates()

# Initialize an empty list to store title history data
title_history_data = []

# Extracting title history
title_columns = ['T1_StartDate', 'T2_StartDate', 'T3_StartDate', 'T4_StartDate', 'T5_StartDate']
for index, row in data.iterrows():
    for idx, title in enumerate(title_columns):
        if pd.notna(row[title]):
            title_history_data.append({
                'ConsultantID': row['ConsultantID'],
                'TitleID': idx + 1,  # TitleID corresponding to the sequence T1, T2, T3, etc.
                'StartDate': row[title]
            })

# Convert title history list to DataFrame
consultant_title_history_table = pd.DataFrame(title_history_data)

# Save the tables to CSV files
consultant_table.to_csv('Consultant.csv', index=False)
consultant_title_history_table.to_csv('Consultant_TitleHistory.csv', index=False)
```
